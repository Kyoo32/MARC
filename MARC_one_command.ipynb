# MARC one-command workflow

MARC is domain- and algorithm-agnostic. Configure the observational time
series, algorithms, five audit axes, and positive controls in one YAML file.

Before running:

1. Put your dataset in `data/`.
2. Copy `marc_config_template.yml` to `marc_config.yml`.
3. Edit `marc_config.yml` for your dataset and study design.

Then execute the complete protocol with the single command below.


```python
from pathlib import Path

CONFIG = Path("marc_config.yml")
if not CONFIG.exists():
    raise FileNotFoundError(
        "Create marc_config.yml from marc_config_template.yml, "
        "then add your dataset and edit the configuration."
    )
print(CONFIG.read_text())
```

```python
# One command: input -> five controlled variations -> anchor -> classify -> report
!python run_marc.py run {CONFIG}
```

```python
import pandas as pd
from IPython.display import display, Image
import yaml

settings = yaml.safe_load(CONFIG.read_text())
output = CONFIG.parent / settings["output"]["directory"]
display(pd.read_csv(output / "robust_core.csv"))
display(pd.read_csv(output / "design_artifacts.csv").head(20))
display(pd.read_csv(output / "positive_control_audit.csv"))
display(Image(filename=output / "figure1_five_axis_profile.png"))
```

## Reuse

Copy `marc_config_template.yml` to `marc_config.yml`, point `input.path` to
your own multivariate time series, and edit the algorithms, audit settings,
and positive controls. No notebook code needs to change.
